```python
#implementing circular queue
class Queue:
    def __init__(self, capacity):
        self.capacity = capacity
        self.front = -1
        self.rear = -1
        self.queue = [None] * capacity

    def enqueue(self, element):
        if (self.rear + 1) % self.capacity == self.front:
            print("Queue is full")
            return
        if self.front == -1:
            self.front = 0
        self.rear = (self.rear + 1) % self.capacity
        self.queue[self.rear] = element
        print(f"{element} inserted into queue")
        print(f"Front: {self.front}")
        print(f"Rear: {self.rear}")

    def dequeue(self):

        if self.front == -1:
            print("Queue is empty")
            return None
        dequeued_element = self.queue[self.front]

        if self.front == self.rear:
            self.front = self.rear = -1
        else:

            self.front = (self.front + 1) % self.capacity

        print(f"{dequeued_element} removed from queue")
        print(f"Front: {self.front}")
        print(f"Rear: {self.rear}")
        return dequeued_element
queue = Queue(5)

queue.enqueue(10)
queue.enqueue(20)
queue.enqueue(30)
queue.enqueue(40)
queue.enqueue(50)

queue.dequeue()
queue.enqueue(60)
queue.dequeue()
```

```python
#anagram
def freq(s):
  dict={}
  for char in s:
    if char in dict:
      dict[char]+=1
    else:
      dict[char]=1
  return dict
def anagram(str1,str2):
  if len(str1)!=len(str2):
    return "not an anagram"
  if freq(str1)==freq(str2):
    return "anagram found"
  else:
    return "not an anagram"
str1="listen"
str2="silent"
print(anagram(str1,str2))
```

```python
#count consecutive 1's
def consecutive(arr):
  count=0
  for i in range(1,len(arr)-1):
    if arr[i]==1:
      if arr[i]==arr[i+1]:
        count+=1
  return count
arr=[2,4,5,1,6,7,2,2,1,1]
print(consecutive(arr))
```
