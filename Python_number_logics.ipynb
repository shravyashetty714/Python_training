```python
def sindig(n):
    while n > 9:  # Continue until n becomes a single-digit number
        sum = 0   # Reset sum for each iteration
        while n > 0:
            sum += n % 10  # Add the last digit
            n //= 10        # Remove last digit
        n = sum  # Update n to the sum for the next iteration
    return n  # Return the final single-digit sum

n = 125369
print(sindig(n))
```
